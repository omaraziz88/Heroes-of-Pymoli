```python
#PURCHASING ANALYSIS (TOTAL):

Unique_Items = len(df['Item ID'].value_counts())
Average_Price = df['Price'].mean()
Sum_Purchases = df['Purchase ID'].count()
Revenue = df['Price'].sum()
df2 = pd.DataFrame(columns = ['Number of Unique Items','Average Purchase Price','Total Number of Purchases','Total Revenue'])
df2.at[0,'Number of Unique Items'] = Unique_Items
df2.at[0,'Average Purchase Price'] = Average_Price
df2.at[0,'Total Number of Purchases'] = Sum_Purchases
df2.at[0,'Total Revenue'] = Revenue

df2.head()
```

```python
#GENDER DEMOGRAPHICS:

#Male_CountNum = (df[df.Gender == 'Male'].count())['Gender'].sum()
#Male_PP = (df[df.Gender == 'Male'].count())['Price'].mean()
Total_Count = df['SN'].nunique()

Filter_M = df['Gender'] == 'Male'
Male_Count = df[Filter_M]['SN'].nunique()
Male_Perc = (Male_Count/Total_Count)*100

Filter_F = df['Gender'] == 'Female'
Female_Count = df[Filter_F]['SN'].nunique()
Female_Perc = (Female_Count/Total_Count)*100

Filter_O = df['Gender'] == 'Other / Non-Disclosed'
Other_Count = df[Filter_O]['SN'].nunique()
Other_Perc = (Other_Count/Total_Count)*100


df3 = pd.DataFrame(columns = ['Total Count','Percentage of Players'], index = ['Male','Female','Other / Non-Disclosed'])
df3.at['Male','Total Count'] = Male_Count
df3.at['Male','Percentage of Players'] = Male_Perc
df3.at['Female','Total Count'] = Female_Count
df3.at['Female','Percentage of Players'] = Female_Perc
df3.at['Other / Non-Disclosed','Total Count'] = Other_Count
df3.at['Other / Non-Disclosed','Percentage of Players'] = Other_Perc

df3.head()
```

```python
#PURCHASING ANALYSIS (GENDER):

Purc_Count_M = df[Filter_M]['Gender'].count()
Purc_Count_F = df[Filter_F]['Gender'].count()
Purc_Count_O = df[Filter_O]['Gender'].count()

Avg_Price_M = df[Filter_M]['Price'].mean()
Avg_Price_F = df[Filter_F]['Price'].mean()
Avg_Price_O = df[Filter_O]['Price'].mean()

Purc_Value_M = df[Filter_M]['Price'].sum()
Purc_Value_F = df[Filter_F]['Price'].sum()
Purc_Value_O = df[Filter_O]['Price'].sum()

PerPerson_M = Purc_Value_M/Male_Count
PerPerson_F = Purc_Value_F/Female_Count
PerPerson_O = Purc_Value_O/Other_Count

df4 = pd.DataFrame({
    'Gender': ['Female','Male','Other / Non-Disclosed'],
    'Purchase Count': [Purc_Count_F,Purc_Count_M,Purc_Count_O],
    'Average Purchase Price': [Avg_Price_F,Avg_Price_M,Avg_Price_O],
    'Total Purchase Value': [Purc_Value_F,Purc_Value_M,Purc_Value_O],
    'Avg Total Purchase per Person': [PerPerson_F,PerPerson_M,PerPerson_O]
})

df4.head()
```
